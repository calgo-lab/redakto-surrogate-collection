```python
from pathlib import Path
from tabulate import tabulate

import collections
import json
import pandas as pd
import re
```

```python
notebook_dir = Path(__file__).parent if "__file__" in globals() else Path.cwd()
csv_path = notebook_dir.parent / "data" / "fakenamegenerator.com" / "FakeNameGenerator.com_ef63c31f.csv"
```

```python
df = pd.read_csv(csv_path, index_col=0)
```

```python
df.head(n=5)
```

```python
column_info = list(zip(df.columns, df.dtypes))
print(tabulate(column_info, headers=["Column Name", "Type"], tablefmt="github"))
```

```python
first_row_values = df.iloc[0].apply(lambda x: x.item() if hasattr(x, "item") else x).tolist()
first_row_info = list(zip(df.columns, first_row_values))
print(tabulate(first_row_info, headers=["Column Name", "Value"], tablefmt="github"))
```

```python
male_given_names = df.loc[df["Gender"] == "male", "GivenName"].unique()
print(f"Total: {len(male_given_names)}\n{male_given_names}")
```

```python
grouped_male_given_names = collections.defaultdict(list)
for name in male_given_names:
    grouped_male_given_names[name[0]].append(name)

for key, value in grouped_male_given_names.items():
    grouped_male_given_names[key] = sorted(grouped_male_given_names[key])

grouped_male_given_names = dict(sorted(grouped_male_given_names.items()))
print(json.dumps(grouped_male_given_names, ensure_ascii=False, indent=4))
```

```python
street_addresses = df.StreetAddress.to_list()
street_names = []
for addr in street_addresses:
    match = re.match(r"^(.+?)\s+\d+$", addr)
    if match:
        street_names.append(match.group(1))
    else:
        street_names.append(addr)
```

```python
street_names = list(set(street_names))
print(f"Total: {len(street_names)}\n{street_names}")
```

```python
grouped_street_names = collections.defaultdict(list)
for name in street_names:
    grouped_street_names[name[0]].append(name)

for key, value in grouped_street_names.items():
    grouped_street_names[key] = sorted(grouped_street_names[key])

grouped_street_names = dict(sorted(grouped_street_names.items()))
print(json.dumps(grouped_street_names, ensure_ascii=False, indent=4))
```
